# Display the count of matches based on the year wise

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pandas AND Metplotib
import pandas as pd
filepath = 'DataFiles/Match.csv'
def readCsvData(filepath):
    return pd.read_csv(filepath)
df = readCsvData(filepath)
df
```

```python
#q1
a=df.groupby('Season_Year')['Country_id'].count()
print(a)
```

# Plot a bar graph  no. of matches won by each team in the year  2011

```python
# q2
a=df[df['Season_Year']==2011]
a
```

```python
a=df[df['Season_Year']==2011]
b=a.groupby('match_winner')['Country_id'].count()
print(b)
```

```python
b=df[df['Season_Year']==2011].groupby('match_winner')['Country_id'].count()
print(b)
q=b.index.tolist()
w=b.values.tolist()
plt.bar(q,w,color='pink',width=0.5)
```
